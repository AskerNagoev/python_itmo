# **Работа с API**

## **Получить информацию обо всех книгах и сохранить в pd.DataFrame**

```python
import json
import requests
import pandas as pd
```

```python
root_url = "https://www.anapioficeandfire.com/api"
```

```python
book_list = []
book_counter = 1
n = 1

while book_counter > 0:
  book_list_page = requests.get(f'{root_url}/books?page={n}&pageSize=50').json()
  book_counter = len(book_list_page)
  book_list.extend(book_list_page)
  n += 1

print("Все книги выгружены")
```

```python
books_df = pd.DataFrame(book_list)
books_df.shape
```

```python
books_df.head(3)
```

## **Получить информацию обо всех домах Вестероса (Houses) и сохранить в pd.DataFrame**

```python
house_list = []
house_counter = 1
n = 1

while house_counter > 0:
  house_list_page = requests.get(f'{root_url}/houses?page={n}&pageSize=50').json()
  house_counter = len(house_list_page)
  house_list.extend(house_list_page)
  n += 1

print("Все дома выгружены")
```

```python
houses_df = pd.DataFrame(house_list)
houses_df.shape
```

```python
houses_df.head(3)
```

## **Получить информацию обо всех домах Вестероса, у которых есть девиз, и сохранить в pd.DataFrame**

```python
house_list_hasTitles = []
house_counter = 1
n = 1

while house_counter > 0:
  house_list_page = requests.get(f'{root_url}/houses?page={n}&pageSize=50&hasTitles=True').json()
  house_counter = len(house_list_page)
  house_list_hasTitles.extend(house_list_page)
  n += 1

print("Все дома выгружены")
```

```python
houses_hasTitle_df = pd.DataFrame(house_list_hasTitles)
houses_hasTitle_df.shape
```

```python
houses_hasTitle_df.head(3)
```
