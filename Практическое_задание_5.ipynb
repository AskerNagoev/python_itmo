# **Оптимизация типов pandas**

## **Загрузка датасета**

```python
mem_before = df.memory_usage(deep=True).sum()
mem_before
```

## **Анализ типов**

```python
df.dtypes
```

```python
# Числовые признаки

df.describe().loc[['min', 'max']]
```

```python
# Категориальные признаки

obj_cols = df.select_dtypes(include='object').columns
for col in obj_cols:
    print(f"{col}: {df[col].unique()}")
```

image.png

*   price и area помещаются в int32
*   bedrooms, bathrooms, stories, parking помещаются в int8
*   mainroad, guestroom, basement, hotwaterheating, airconditioning, prefarea, у которых бинарные значения эффективнее хранить как bool
*   furnishingstatus имеет три значения, можно заменить их тоже

## **Изменение типов**

```python
# Оптимизация числовых признаков

df['price'] = df['price'].astype('int32')
df['area'] = df['area'].astype('int32')

cols = ['bedrooms', 'bathrooms', 'stories', 'parking']
df[cols] = df[cols].astype('int8')
```

```python
df.dtypes
```

```python
# Категориальные признаки

cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[cols] = df[cols].replace({'yes': True, 'no': False}).astype('bool')

df['furnishingstatus'] = df['furnishingstatus'].astype('category')
```

```python
df.dtypes
```

```python
mem_after = df.memory_usage(deep=True).sum()
```

```python
# ДО
mem_before/1024
```

```python
# После
mem_after/1024
```
