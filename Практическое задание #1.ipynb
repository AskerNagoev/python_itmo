```python
# Создание класса
class Chocolate:
  def __init__(self, price, additive=None): # Характеристики объекта
    if float(price) <= 0: # Проверка стоимости
      raise ValueError('Цена равна нулю или меньше')
    self.price = float(price)
    self.additive = additive

  def show_my_chocolate(self):  # Метод вывода добавки
    if self.additive is not None:
      print(f'Шоколадный батончик, добавка: {self.additive}')
    else:
      print('Обычный шоколадный батончик')

  def __str__(self): # Преобразование объекта класса в строку
    return f'Шоколадный батончик, {self.price} рублей'
```

```python
# Подготовка данных
price = ['30', 35, '45', 50]
additive = ['Nuts', 'Caramel', 'Strawberry', 'Popcorn']
```

```python
# Создание экземпляров
chocos = []
for price_val, additive_val in zip(price, additive):
  chocos.append(Chocolate(price_val, additive_val))
```

```python
# Вывод строк
for item in chocos:
  print(item)
```

```python
# Вывод добавок
for item in chocos:
  item.show_my_chocolate()
```

```python
# Создание экземпляра без добавки и его вывод
choco = Chocolate(25)
print(choco)
choco.show_my_chocolate()
```

```python
# Создание экземпляра со стоимостью меньше 0
choco = Chocolate(-2)
print(choco)
choco.show_my_chocolate()
```
